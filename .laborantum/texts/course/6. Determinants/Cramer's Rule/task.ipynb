\end{cases}$

\end{cases}$

x_1 + x_2 + x_3 = 2 \\

x_1 + 2 x_2 + 2 x_3 = 3
\end{cases}$

2 x_1 + x_2 + x_3 = 1 \\
x_1 + x_2 = 0 \\
